**Plot a chart for each column**

Các cột hành vi người dùng trên website

```python
import seaborn as sns
import os
columns = ['Administrative', 'Informational', 'ProductRelated']
for column in columns:
    plt.figure(figsize=(8, 5))
    sns.histplot(df[column], bins=10, edgecolor='k', kde=False)
    
    clean_col = column.replace('_', ' ')
    plt.title(f'Biểu đồ tần suất: {clean_col}', fontsize=14, fontweight='bold')
    plt.xlabel(f'Số lần truy cập trang {clean_col}', fontsize=12)
    plt.ylabel('Số lượng phiên', fontsize=12)
    plt.grid(axis='y', linestyle='--', alpha=0.5)
```

```python
columns = ['Administrative_Duration','Informational_Duration', 'ProductRelated_Duration']
# Lặp qua từng cột của DataFrame và vẽ histogram cho các cột kiểu nguyên
for column in columns:
    plt.figure(figsize=(8, 5))
    sns.histplot(df[column], bins=10, edgecolor='k', kde=False)
    
    # Đổi tên biểu đồ & nhãn trục
    clean_col = column.replace('_', ' ')
    plt.title(f'Biểu đồ tần suất: {clean_col}', fontsize=14, fontweight='bold')
    plt.xlabel(f'Thời gian truy cập trang {clean_col} (s)', fontsize=12)
    plt.ylabel('Số lượng phiên', fontsize=12)
    plt.grid(axis='y', linestyle='--', alpha=0.5)
```

Boxplot (Revenue with duration time)

```python
import pandas as pd
import matplotlib.pyplot as plt

duration_cols = [
    'Administrative_Duration',
    'Informational_Duration',
    'ProductRelated_Duration'
]
for col in duration_cols:
    plt.figure(figsize=(6, 4))
    df.boxplot(column=col, by='Revenue')
    plt.title(f'Thời gian trên trang {col.replace("_Duration", "")} theo trạng thái mua hàng')
    plt.suptitle('')  # Xóa tiêu đề phụ
    plt.xlabel('Trạng thái mua hàng(Revenue)')
    plt.ylabel('Thời gian (giây)')
    plt.show()
```

2.Cột hành vi thoát trang

```python
rate_cols = ['BounceRates', 'ExitRates']

for col in rate_cols:
    plt.figure(figsize=(6, 4))
    df.boxplot(column=col, by='Revenue')
    if col == 'BounceRates':
        col = 'Tỉ lệ thoát'
    else:
        col = 'Tỉ lệ thoát cuối'
    plt.title(f'{col} theo trạng thái mua hàng')
    plt.suptitle('')  
    plt.xlabel('Trạng thái mua hàng (Revenue)')
    plt.ylabel('Tỷ lệ')
    plt.show()
```

3. Giá trị trang và ngày đặc biệt

```python
column = 'PageValues'

plt.figure(figsize=(8, 5))
sns.histplot(df[column], bins=10, edgecolor='k', kde=False)

plt.title(f'Biểu đồ tần suất: {column}', fontsize=14, fontweight='bold')
plt.xlabel('Giá trị trang', fontsize=12)
plt.ylabel('Số lượng phiên', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.show()
```

```python
def categorize_special(x):
    if x == 0:
        return 'NoEvent'
    elif x == 1:
        return 'OnEvent'
    else:
        return 'NearEvent'

df['SpecialDay_cat'] = df['SpecialDay'].apply(categorize_special)


categories = ['NoEvent', 'NearEvent', 'OnEvent']
df['SpecialDay_cat'] = pd.Categorical(df['SpecialDay_cat'],
                                      categories=categories,
                                      ordered=True)

# Tính số phiên cho từng nhóm
grouped = df.groupby(['SpecialDay_cat', 'Revenue']).size().unstack(fill_value=0)

# Vẽ grouped bar chart
ax = grouped.plot(kind='bar', figsize=(8, 5))

ax.set_title('Biểu đồ thể hiện mối tương quan giữa ngày đặc biệt với trạng thái mua hàng', fontsize=14, fontweight='bold')
ax.set_xlabel('SpecialDay Category', fontsize=12)
ax.set_ylabel('Number of Sessions', fontsize=12)
ax.legend(title='Revenue')
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.show()
```
