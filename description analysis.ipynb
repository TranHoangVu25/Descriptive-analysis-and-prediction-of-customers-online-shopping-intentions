Boxplot (Revenue with duration time)

```python
import pandas as pd
import matplotlib.pyplot as plt

duration_cols = [
    'Administrative_Duration',
    'Informational_Duration',
    'ProductRelated_Duration'
]
for col in duration_cols:
    plt.figure(figsize=(6, 4))
    df.boxplot(column=col, by='Revenue')
    plt.title(f'Thời gian trên trang {col.replace("_Duration", "")} theo trạng thái mua hàng')
    plt.suptitle('')  # Xóa tiêu đề phụ
    plt.xlabel('Trạng thái mua hàng(Revenue)')
    plt.ylabel('Thời gian (giây)')
    plt.show()
```

3. Giá trị trang và ngày đặc biệt

```python
column = 'PageValues'

plt.figure(figsize=(8, 5))
sns.histplot(df[column], bins=10, edgecolor='k', kde=False)

plt.title(f'Biểu đồ tần suất: {column}', fontsize=14, fontweight='bold')
plt.xlabel('Giá trị trang', fontsize=12)
plt.ylabel('Số lượng phiên', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.show()
```

```python
def categorize_special(x):
    if x == 0:
        return 'Không có sự kiện'
    elif x == 1:
        return 'Có sự kiện'
    else:
        return 'Gần sự kiện'

df['SpecialDay_cat'] = df['SpecialDay'].apply(categorize_special)


categories = ['Không có sự kiện', 'Gần sự kiện', 'Có sự kiện']
df['SpecialDay_cat'] = pd.Categorical(df['SpecialDay_cat'],
                                      categories=categories,
                                      ordered=True)

# Tính số phiên cho từng nhóm
grouped = df.groupby(['SpecialDay_cat', 'Revenue']).size().unstack(fill_value=0)

# Vẽ grouped bar chart
ax = grouped.plot(kind='bar', figsize=(8, 5))

ax.set_title('Biểu đồ thể hiện mối tương quan giữa ngày đặc biệt với trạng thái mua hàng', fontsize=14, fontweight='bold')
ax.set_xlabel('Ngày đặc biệt', fontsize=12)
ax.set_ylabel('Số lượng phiên', fontsize=12)
ax.legend(title='Trạng thái mua')
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.show()
```

4. Thông tin hệ thống & hành vi người dùng

count plot OperatingSystems, Browser, Region, TrafficType

```python
encode_cols = ['OperatingSystems', 'Browser', 'Region', 'TrafficType']

for col in encode_cols:
    plt.figure(figsize=(8, 5))
    sns.countplot(
        data=df,
        x=col,
        hue='Revenue',
        order=sorted(df[col].unique())
    )
    if col == 'OperatingSystems':
        col = 'Hệ điều hành'
    elif col == 'Browser':
        col = 'Trình duyệt web'
    elif col == 'Region':
        col = 'Quốc gia'
    elif col == 'TrafficType':
        col = 'Nguồn truy cập'     
    plt.title(f'Biểu đồ thể hiện mối quan hệ của {col} với trạng thái mua hàng', fontsize=14, fontweight='bold')
    plt.xlabel(col)
    plt.ylabel('Số phiên')
    plt.legend(title='Revenue')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

```python
heatmap_data = pd.crosstab(df['TrafficType'], df['Revenue'])

# Vẽ heatmap
plt.figure(figsize=(10,6))
sns.heatmap(heatmap_data, annot=True, cmap='YlGnBu', fmt='d')
plt.title('Biểu đồ tần suất thể hiện mối quan hệ giữa loại phương tiện và trạng thái mua hàng')
plt.ylabel('Loại phương tiện')
plt.xlabel('Trạng thái mua hàng')
plt.show()
```
