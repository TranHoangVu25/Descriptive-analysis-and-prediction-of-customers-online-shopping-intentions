```python
import seaborn as sns
import os
df = pd.read_csv('online_shoppers_intention.csv')

# Lp qua tng ct ca DataFrame v v histogram cho cc ct kiu nguyn
for column in df.select_dtypes(include=['int64']):
    plt.hist(df[column], bins=10, edgecolor='k')
    plt.title(f'Histogram of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    file_path = os.path.join('Histogram', f'{column}_histogram.png')
    # plt.savefig(file_path)
    # plt.close() 
    plt.show()



```

```python
#  v biu  hp, ta cn a d liu v dng s
#  Chn cc ct d liu s (numeric columns)
df_numeric = df.select_dtypes(include=['number'])

for column in df_numeric.columns:
    plt.figure(figsize=(6, 4))  # Kch thc ca biu 
    plt.boxplot(df[column])
    plt.title(f'Box Plot of {column}')
    plt.ylabel(column)
    plt.grid(True)
    # file_path = os.path.join('Box Plot', f'{column}_box_plot.png')
    # plt.savefig(file_path)
    # plt.close() 

    plt.show()


```
